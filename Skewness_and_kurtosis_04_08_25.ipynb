```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

import datetime
from datetime import date, timedelta
```

```python
df=yf.download('TCS.NS',
               start='2019-01-01',
               end="2025-01-01")
```

```python
df['Close'].plot(figsize=(16,6)).axhline(y=4060,color='r',linestyle='--',label='reference line')
plt.legend()
plt.show()
```

```python
df["Close_simple_return"]=df["Close"].pct_change()
```

```python
df["Close_simple_return"]
```

```python
df
```

```python
df["Close_simple_return"].plot(figsize=(10,7))
```

```python
df["Close_simple_return"].plot(figsize=(10,7)).axhline(y=0,color='r',linestyle='--',label='reference line')
plt.legend()
```

```python
df["Close_simple_return"].describe()
```

```python
df["Close_simple_return"].corr(df["Close_simple_return"].shift(1))
```

```python
#calcutate skewness and kurtosis
skewness=df["Close_simple_return"].skew()
kurtosis=df["Close_simple_return"].kurtosis()
print("Skewness:",skewness)
print("Kurtosis:",kurtosis)
```

```python
skewness1=df["Close"].skew()
kurtosis1=df["Close"].kurtosis()
print("Skewness:",skewness1)
print("Kurtosis:",kurtosis1)
```
